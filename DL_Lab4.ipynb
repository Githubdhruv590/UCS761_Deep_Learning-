```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/content/multiple_linear_regression_dataset.csv")

print(data.head())
print(data.columns)
print(data.shape)
```

The dataset contains 3 columns:
- age
- experience
- income

There are 20 rows and 3 columns.

The input features are age and experience.
The output (target variable) is income.

So the model needs to learn 2 weights and 1 bias.


```python
X = data[["age", "experience"]].values
y = data["income"].values

print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

X has 2 columns because we have 2 features.

y has only one column because income is a single numeric output.

```python
n_features = X.shape[1]

w = np.zeros(n_features)
b = 0.0

print("Initial weights:", w)
print("Initial bias:", b)
```

We initialize one weight per feature because each input
(age and experience) should have its own influence on income.

Bias is kept separate because it represents the base income
when input features are zero.

Initializing with very large values would be risky because
it could produce very large predictions,
leading to huge errors and unstable training.


```python
def predict(X, w, b):
    return X.dot(w) + b
```

This step defines how the model makes predictions.

It computes a linear combination of input features
using weights and bias.

There is no activation function because this is a regression task.
We want to predict a continuous numeric value (income).

y_hat can take any real value.

This is different from logistic regression,
where the output is passed through sigmoid
and restricted between 0 and 1.

```python
def mean_squared_error(y, y_hat):
    return ((y_hat - y) ** 2).mean()
```

We square the error so that negative and positive differences
do not cancel each other.

If one prediction is very wrong,
its squared error becomes large
and increases total loss significantly.

We do not use absolute error because
squared error is smoother and easier to optimize.


```python
def compute_gradients(X, y, y_hat):
    N = len(y)

    dw = (2 / N) * X.T.dot(y_hat - y)
    db = (2 / N) * (y_hat - y).sum()

    return dw, db
```

X appears in dw because weights multiply input features but bias does not hence not in db.

The error term appears everywhere because learning depends on how wrong the prediction is.

If error becomes zero then learning stops because gradient will be zero.


```python
def update_parameters(w, b, dw, db, lr):
    w = w - lr * dw
    b = b - lr * db
    return w, b
```

```python
lr = 0.0001
epochs = 1000

for epoch in range(epochs):
    y_hat = predict(X, w, b)

    loss = mean_squared_error(y, y_hat)

    dw, db = compute_gradients(X, y, y_hat)

    w, b = update_parameters(w, b, dw, db, lr)

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")
```

Yes, the loss decreases over time.
It starts very large and gradually becomes smaller.

If loss increases, it usually means the learning rate is too high,
and the model is taking steps that are too big.

Learning rate and epochs work together.
A small learning rate needs more epochs.
A large learning rate may reduce loss faster,
but it can also make training unstable.

```python
new_candidate = np.array([30, 6])

predicted_salary = new_candidate.dot(w) + b

print("Predicted Salary:", predicted_salary)
print("Final Weights:", w)
print("Final Bias:", b)
```

The predicted income looks reasonable.

Model changes output smoothly when age or experience changes.
So interpolation is happening properly.

This is better than threshold rules
because income is continuous value,
not yes or no type output.
