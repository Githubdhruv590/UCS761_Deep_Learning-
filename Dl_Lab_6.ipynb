```python
import numpy as np
np.random.seed(42)
```

```python
X = np.linspace(0, 10, 50).reshape(-1, 1)

print(X.shape)
X
# X represents a single input feature.
# It can be thought of as some quantity increasing over time.
# The range 0 to 10 is chosen to capture
# fast early change and slower later change.
```

```python
noise = np.random.normal(0, 0.2, size=(50, 1))
y = np.log(X + 1) + noise

print(y.shape)
y
# The target grows fast initially and slows down later.
# This logarithmic shape cannot be fit well by a straight line.
# A linear model has only one slope and cannot bend.
```

```python
# We choose 3 hidden units.
# More than 1 hidden unit allows the model to bend.
# Too many hidden units can make learning unstable.
# 3 is a balance between flexibility and simplicity.
```

```python
W1 = np.random.uniform(-1, 1, size=(1, 3))
b1 = np.zeros((1, 3))

W2 = np.random.uniform(-1, 1, size=(3, 1))
b2 = np.zeros((1, 1))
# W1 controls how input is transformed into hidden units.
# b1 shifts hidden activations left or right.
# W2 controls how hidden units combine to form output.
# b2 shifts final output up or down.
# Weights affect shape, biases affect position.
```

```python
def activation(z):
    return np.maximum(0, z)
```

```python
def activation_slope(z):
    return (z > 0).astype(float)
# Without slope information, the model would not know how to change parameters.
```

```python
z1 = X @ W1 + b1
h = activation(z1)
y_hat = h @ W2 + b2
```

```python
error = y_hat - y


loss = np.mean(error ** 2)

print("Initial loss:", loss)
# Squaring makes large errors more expensive.
# This forces the model to correct big mistakes.
```

```python
# To reduce loss, parameters must change.
# To know how to change them, we need to know whether loss increases or decreases when parameters change.
# This is why gradients (slopes) are required.
```

```python
dL_dy = 2 * error / len(X)


dL_dW2 = h.T @ dL_dy
dL_db2 = np.sum(dL_dy, axis=0, keepdims=True)


dL_dh = dL_dy @ W2.T


dL_dz1 = dL_dh * activation_slope(z1)
# Activation slope controls where error can pass.
# If slope is zero, error is blocked.

dL_dW1 = X.T @ dL_dz1
dL_db1 = np.sum(dL_dz1, axis=0, keepdims=True)
```

```python
learning_rate = 0.01

W1 -= learning_rate * dL_dW1
b1 -= learning_rate * dL_db1
W2 -= learning_rate * dL_dW2
b2 -= learning_rate * dL_db2
# This is same idea as perceptron update:
# move parameters opposite to error direction.
```

```python
epochs = 1000

for epoch in range(epochs):
    # Forward pass
    z1 = X @ W1 + b1
    h = activation(z1)
    y_hat = h @ W2 + b2

    # Loss
    error = y_hat - y
    loss = np.mean(error ** 2)

    # Backward pass
    dL_dy = 2 * error / len(X)
    dL_dW2 = h.T @ dL_dy
    dL_db2 = np.sum(dL_dy, axis=0, keepdims=True)

    dL_dh = dL_dy @ W2.T
    dL_dz1 = dL_dh * activation_slope(z1)

    dL_dW1 = X.T @ dL_dz1
    dL_db1 = np.sum(dL_dz1, axis=0, keepdims=True)

    # Update
    W1 -= learning_rate * dL_dW1
    b1 -= learning_rate * dL_db1
    W2 -= learning_rate * dL_dW2
    b2 -= learning_rate * dL_db2

    if epoch % 100 == 0:
        print(epoch, loss)
# Learning rate is chosen small for stability.
# 1000 epochs allow gradual learning.
# Loss should decrease as model bends to fit data.
```
